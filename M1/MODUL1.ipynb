```python
import matplotlib.pyplot as plt

# Buat canvas
plt.figure(figsize=(5, 5))
plt.axis('off')

# Koordinat persegi panjang
x1, x2 = 2, 8
y1, y2 = 2, 6

# Gambar empat sisi dengan warna berbeda
plt.plot([x1, x2], [y2, y2], 'r-')  # atas - merah
plt.plot([x1, x2], [y1, y1], 'b-')  # bawah - biru
plt.plot([x1, x1], [y1, y2], 'g-')  # kiri - hijau
plt.plot([x2, x2], [y1, y2], 'm-')  # kanan - ungu

# Tampilkan gambar
plt.show()

```

```python
import matplotlib.pyplot as plt

# Tentukan ukuran canvas
canvas_width = 10
canvas_height = 10

# Membuat canvas kosong dengan latar belakang putih
canvas = plt.figure(figsize=(canvas_width, canvas_height))
ax = canvas.add_subplot(111)

# Matikan axis (sumbu)
ax.axis('off')

# Gambar titik piksel di koordinat (5, 5)
ax.plot(5, 5, 'ro')  # 'ro' berarti titik merah

# Tampilkan gambar
plt.show()
```

```python
import matplotlib.pyplot as plt

# Ukuran canvas
canvas_width = 10
canvas_height = 10

# Buat canvas
fig = plt.figure(figsize=(canvas_width, canvas_height))
ax = fig.add_subplot(111)
ax.axis('off')  # Matikan sumbu

# Koordinat persegi panjang
x_min, x_max = 2, 8
y_min, y_max = 2, 6

# Titik sisi atas (merah)
for x in range(x_min, x_max + 1):
    ax.plot(x, y_max, 'ro')  # 'ro' = titik merah

# Titik sisi bawah (hijau)
for x in range(x_min, x_max + 1):
    ax.plot(x, y_min, 'go')  # 'go' = titik hijau

# Titik sisi kiri (biru)
for y in range(y_min, y_max + 1):
    ax.plot(x_min, y, 'bo')  # 'bo' = titik biru

# Titik sisi kanan (kuning)
for y in range(y_min, y_max + 1):
    ax.plot(x_max, y, 'yo')  # 'yo' = titik kuning

# Tampilkan hasil
plt.show()
```
